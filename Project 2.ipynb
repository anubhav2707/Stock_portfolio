```python
import pandas as pd
import numpy as np
from queue import Queue
 
```

```python
temp=pd.read_csv("project2.csv")
df1=pd.DataFrame(temp)
dfn=df1.T
df2=dfn.convert_dtypes()
df=df2.copy()
df.head(5)
```

```python
#function to calculate returns 
for j in range(len(df)):
    for i in range(1,len(df)-1):
            b=df.iloc[i][j]-df.iloc[i+1][j]
            df.iloc[i, df.columns.get_loc(df.columns[j])]=b
            
            
      
```

```python
df.columns = df.iloc[0]
df.drop(df.index[0],inplace=True)
df.dropna(inplace=True)
df.head(5)
```

```python
for j in range(1,len(df.columns)):
    df[j]=df.iloc[:,j].rolling(window=9).mean()
df.dropna(axis='columns',inplace=True)
df.head(5)
```

```python
df.index = pd.to_datetime(df.index)
df.sort_index(axis=0, ascending=True, inplace=True)
df.drop(df.index[0],inplace=True)
df
```

```python
#function for sorting of list
            
def Sort(sub_li): 
    l = len(sub_li) 
    for i in range(0, l): 
        for j in range(0, l-i-1): 
            if (sub_li[j][1] > sub_li[j + 1][1]): 
                tempo = sub_li[j]  
                sub_li[j]= sub_li[j + 1] 
                sub_li[j + 1]= tempo 
    return sub_li             
    
```

```python
top20=[]
final=[]
lis=[]
flag=0
weight=[]
for row in df.index: 
    list1=[]
    top10=[]
    finale=[]
    vol=[]
    weight_temp=[]
    df.sort_values(by=[row],axis=1,ascending=False,inplace=True)
   
    for i in range(0,10):
        if flag==0:
            lis.append(df.columns[i])
            continue
        found=0
        #loop for ensuring that only those stocks comes in list1 which weren't there in lis
        for j in range(0,len(lis)):
            if lis[j]==df.columns[i]:
                found=1
        if found==0:    
            list1.append(df.columns[i])
    
    if flag==0:
        final.append(lis)
        for i in range(0,20):
            top20.append(df.columns[i])
    
    else: 
        pos=-2
        for i in range(0,len(lis)):
            mark=0
            for k in range(0,len(df.columns)):
                if df.columns[k]==lis[i]:
                    pos=k
                    break
            if i<=4:
                top10.append([lis[i],pos])
            else:
                for j in range(0,20):
                    if top20[j]==lis[i]:
                        top10.append([lis[i],pos])
                        mark=1
                        break
                        
                if mark==0:
                    for k in range(0,len(df.columns)):
                        if df.columns[k]==list1[0]:
                            pos=k
                            break
                    top10.append([list1[0],pos])
                    list1.pop(0)
       
        top10=Sort(top10)
        
        for i in range(0,10):
            finale.append(top10[i][0])
            
        for i in range(0,20):
            top20.pop(0)
            top20.append(df.columns[i])
        
        
        final.append(finale)
        
        #emptyig the lis and filling it up with the 10 new stocks
        while len(lis)!=0:
            lis.pop(0)
        
        for i in range(0,10):
            lis.append(finale[i])
        
        
        #now calculating volatility and giving them weightage

        for i in range(0,10):
            
            a=(np.std(df[finale[i]])).astype(int)
     
            vol.append([i,a])
            
        vol1=Sort(vol)
        
        for i in range(0,10):
            weight_temp.append([finale[vol1[i][0]],(10-i)*1.8181])
        weight.append(row)
        weight.append(weight_temp)
            
            
    flag=1  #displaying stock with their weightage
weight
```
